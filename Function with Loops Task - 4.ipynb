```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        amount = int(input("Enter recharge amount: ₹"))
        if amount < 50:
            print("Minimum recharge should be ₹50")
            continue
        if amount not in valid_plans:
            print(" Invalid plan. Choose from:", valid_plans)
            continue
        print(" Recharge Successful!")
        break
# Call function
validate_recharge()
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(stock):
    for product, qty in stock.items():
        if qty < 15:
            print(f" Reorder Alert: {product} stock is low ({qty})")
        else:
            print(f" Stock OK: {product} ({qty})")
# Sample data
inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 5,
    "Milk": 20
}
check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System
def check_result(marks):
    total = 0
    for m in marks:
        total += m
    avg = total / len(marks)
    print("Average Marks:", avg)
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"
marks_list = [60, 55, 70, 40]
result = check_result(marks_list)
print("Result:", result)
```

```python
#Problem 4: Cab Fare Estimator with Retry
def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)
    if peak:
        fare = fare + (fare * 0.25)
    return fare
while True:
    km = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()
    peak = True if peak_input == "yes" else False
    total_fare = calculate_fare(km, peak)
    print("Total Fare: ₹", total_fare)
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def check_attendance(attendance_list):
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1
    percentage = (present_days / len(attendance_list)) * 100
    print("Attendance %:", percentage)
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance = ["P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance)
print("Status:", status)
```

```python
#Problem 6: Password Strength Checker
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    if len(password) < 8:
        return "Weak Password (Too short)"
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
pwd = input("Enter password: ")
print(check_password(pwd))
```
