```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Alisha": 92,
    "Aarti": 88,
    "Isha": 92,
    "Sahil": 85
}

# Step 1: Find highest score
max_score = max(employees.values())

# Step 2: Find employees with highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == max_score]

# Step 3: Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation (basic handling)
import string
query = query.translate(str.maketrans("", "", string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Step 5: Filter words with frequency > 1
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Step 1: Loop through readings with index
for index, value in enumerate(sensor_readings):
    # Step 2: Check if even
    if value % 2 == 0:
        valid_readings.append((index, value))

# Step 3: Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "alisha@gmail.com",
    "aarti@yahoo.com",
    "isha@gmail.com",
    "sahil@gmail.com",
    "kiran@yahoo.com"
]

domain_count = {}

# Step 1: Extract domains and count
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate total emails
total = len(emails)

# Step 3: Display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average
average_sales = sum(sales) / len(sales)

# Step 2: Define spike threshold (30% above average)
threshold = average_sales * 1.3

# Step 3: Find spike days
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Step 1: Count occurrences
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Filter duplicates
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
